# Combining dataframes and getting statistics

The goal of this notebook is to fetch both the original database we have been working from, and the results we fetched with the API (the average price of the top 100 board games).  
Once we have them in a combined database, we can start creating some relevant statistics and plots.

```python
import pandas as pd
import requests
import os
import statistics as stat
import functions as s
import matplotlib as m

df = pd.read_csv("../input/topBGs2019")
df.head()
```

```python
dfp = pd.read_csv("../output/top100price.csv", names=['names','avgPrice'])
dfp.head()
```

```python
dfc = df.merge(right = dfp, on = 'names', how = 'left')
dfc.head()
top100 = dfc[:100]
```

```python
dfplayers = dfc.groupby(['max_players']).names.count()

# Graph of number of max players by count of games
maxplayers_by_count = dfplayers.plot(xlim=[0,20]).set_ylabel('game count')
```

```python
avgprice100 = top100.mean()['avgPrice']
medianprice100 = top100.median()['avgPrice']
avgplaytime = dfc.mean()['avg_time']
medianplaytime = dfc.median()['avg_time']
```
